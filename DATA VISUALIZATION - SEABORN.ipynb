# Seaborn visualization is better than matplotlib. 
# So we can read data perfectly

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
data=pd.read_csv("train.csv")
```

# SCATTER PLOT

```python
sns.scatterplot(x="Age",y="Fare",hue="Survived",data=data); # hue 2 dimensional graphic made in 3 dimensions
                                                            # data=data indicates the source of the data.
```

# LINE PLOT

```python
sns.lineplot(data=data[["Age","Fare"]]); 
```

# HİSTOGRAM / DİSTRUBUTİON PLOT

```python
sns.distplot(data["Age"],bins=10,kde=True); # bins--> 10 20 30 40  / kde=True if line is avilable.
```

# BAR CHART/COUNTPLOT

```python
sns.countplot(data["Embarked"]);
```

# BOXPLOT

```python
sns.boxplot("Age","Sex",data=data);
```

# HEATMAP 

```python
sns.heatmap(data.corr(),annot=True);# Correlation
```
